```python
import cv2

img = cv2.imread("../data/raw/blood_cell.jpg")
print(img.shape)  
```

```python
cv2.imshow("My Image", img)
cv2.waitKey(0)      
cv2.destroyAllWindows()
```

```python
resized = cv2.resize(img, (640, 480))
print(resized.shape) 
```

```python
# OpenCV loads images in BGR by default (not RGB!)
# Convert BGR → RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert RGB → BGR
bgr = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)  # (height, width) — no channel dimension!
```

```python
# cv2.rectangle(image, top-left corner, bottom-right corner, color, thickness)
cv2.rectangle(img, (50, 50), (200, 200), (0, 255, 0), 2)

# Add a label
cv2.putText(img, "RBC", (50, 45), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

cv2.imshow("Bounding Box", img)
cv2.waitKey(0)
```

```python
# img[y1:y2, x1:x2]
cropped = img[50:200, 50:200]

cv2.imshow("Cropped", cropped)
cv2.waitKey(0)
```
